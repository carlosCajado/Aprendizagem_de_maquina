```python
import pandas as pd
import numpy   as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import zero_one_loss, classification_report
sns.set_theme()
Dataset = pd.read_csv('wdbc_normalizacao.csv')
#Dataset.round(2)
Dataset
```

```python
from sklearn.model_selection import train_test_split

# Dividindo em treino e teste 85% treino e 15% teste:
D_treino, D_teste = train_test_split(Dataset, test_size=.15, random_state=30) #test_size, por padrão é 25%

print("D_treino: ", D_treino.shape, dict(D_treino.iloc[:, 0].value_counts()))
print("D_teste: ", D_teste.shape, dict(D_teste.iloc[:, 0].value_counts()))
```

```python
# visualizar as amostras:
plt.figure(figsize=(4,3))
plt.hist([D_treino.iloc[:, 0], D_teste.iloc[:, 0]], bins=[0,1,2], align='left', rwidth=0.5)
plt.legend(['D_treino','D_teste'])
plt.xlabel('Y'), plt.xticks([0,1])
plt.show()
```

```python
# Selecionado os atributos e classes de treino e teste.
X_treino = D_treino.iloc[:, 1:31]
Y_treino = D_treino.iloc[:, 0]
print("D_treino: ", (X_treino.shape, Y_treino.shape))

X_teste = D_teste.iloc[:, 1:31]
Y_teste = D_teste.iloc[:, 0]
print("D_teste: ", (X_teste.shape, Y_teste.shape))
```

**Normalização:**

```python
def Normalization(x):
        
    return x/x.abs().max()
```

```python
X_treino_Normalizado = Normalization(X_treino)
X_teste_Normalizado = Normalization(X_teste)
```

```python
# Importando o KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando o classificador
clf = KNeighborsClassifier(n_neighbors=2)
```

```python
# Fazendo o fit com os dados de treino
clf = clf.fit(X_treino_Normalizado,Y_treino)
```

```python
# Fazendo a previsão para os dados de teste
y_pred = clf.predict(X_teste_Normalizado)
```

```python
# Verificando a matriz de confusão
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_teste,y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_teste, y_pred)
print("Acurácia ", accuracy)
```

```python
def teste(g, D_tst, metrica, **kwds):
    X_tst, y_tst = D_tst
    y_true = y_tst
    y_pred = clf.predict(X_tst)
    E_tst = metrica(y_true, y_pred, **kwds)
    
    return E_tst


def validacao(Modelo, D_trn, D_val, metrica, **kwds):
    # Treinamento:
    g = Modelo.fit(*D_trn)
    
    # Teste:
    E_trn = teste(g, D_trn, metrica, **kwds)
    E_val = teste(g, D_val, metrica, **kwds)
    
    return E_trn, E_val
```

```python
# Dividindo em  dados de treino e validação 15%
X_train, X_val, y_train, y_val = train_test_split(X_treino_Normalizado, Y_treino, test_size=0.15, random_state=30)
```

```python
#Avaliando o modelo: 
E_trn, E_tst = validacao(clf, (X_train, y_train), (X_teste, Y_teste), zero_one_loss)
print("Erro de treinamento:", E_trn)
print("Erro de teste :", E_tst)
```

```python
# Plot matriz de confusão
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm_trn, cm_tst = validacao(clf, (X_train, y_train), (X_teste_Normalizado, Y_teste), confusion_matrix)

with sns.plotting_context(rc={'grid.linewidth': 0}):
    fig, ax = plt.subplots(ncols=2, figsize=(5,2))
    ConfusionMatrixDisplay(confusion_matrix=cm_trn, display_labels=[0,1]).plot(ax=ax[0], colorbar=False)
    ConfusionMatrixDisplay(confusion_matrix=cm_tst, display_labels=[0,1]).plot(ax=ax[1], colorbar=False)
```

```python
# Avaliação modelo  (relatório de classificação)
cr_trn, cr_tst = validacao(clf, (X_train, y_train), (X_teste_Normalizado, Y_teste),
                            classification_report, target_names=[' 0', '1'])
print("treino:\n", cr_trn)
print("teste :\n", cr_tst)
```

```python
# Variando o valor de k de 1 a 10
k_values = range(1, 11)
accuracies = []

for k in k_values:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_treino_Normalizado, Y_treino)
    y_pred = clf.predict(X_teste_Normalizado)
    accuracy = accuracy_score(Y_teste, y_pred)
    accuracies.append(accuracy)
    print(f"k={k}, Acurácia={accuracy:.4f}")

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('Número de Vizinhos k')
plt.ylabel('Acurácia')
plt.title('Acurácia do k-NN Variando o Valor de k')
plt.grid(True)
plt.show()
```

```python
def avaliar_knn(k_values, distance_metric, use_weights=False):
    accuracies = []

    for k in k_values:
        if use_weights:
            clf = KNeighborsClassifier(n_neighbors=k, metric=distance_metric, weights='distance')
        else:
            clf = KNeighborsClassifier(n_neighbors=k, metric=distance_metric)
        
        clf.fit(X_treino_Normalizado, Y_treino)
        y_pred = clf.predict(X_teste_Normalizado)
        accuracy = accuracy_score(Y_teste, y_pred)
        accuracies.append(accuracy)
        print(f"k={k}, Acurácia={accuracy:.4f}, Metrica={distance_metric}, use_weights={use_weights}")
        
        # Analisar classificações incorretas
        incorrect_indices = np.where(y_pred != Y_teste)[0]
        if len(incorrect_indices) > 0:
            print(f"Classificações incorretas para k={k}:")
            for idx in incorrect_indices:
                print(f"Índice: {idx}, Classe Real: {Y_teste.iloc[idx]}, Predição: {y_pred[idx]}")
        
        # Mostrar matriz de confusão e relatório de classificação para k 
        if k == 2:
            print(f"\nMatriz de Confusão para k={k}:\n{confusion_matrix(Y_teste, y_pred)}")
            print(f"\nRelatório de Classificação para k={k}:\n{classification_report(Y_teste, y_pred)}")

    return accuracies
```

```python
k_values = range(1, 11)

# Distância Euclidiana sem pesos
accuracies_euclidean = avaliar_knn(k_values, distance_metric='euclidean')
```

```python
# Distância Euclidiana com pesos
accuracies_euclidean_weights = avaliar_knn(k_values, distance_metric='euclidean', use_weights=True)
```

```python
# Distância Manhattan sem pesos
accuracies_manhattan = avaliar_knn(k_values, distance_metric='manhattan')
```

```python
# Distância Manhattan com pesos
accuracies_manhattan_weights = avaliar_knn(k_values, distance_metric='manhattan', use_weights=True)
```

```python

# resultados
plt.figure(figsize=(12, 8))

plt.plot(k_values, accuracies_euclidean, marker='o', label='Euclidiana')
plt.plot(k_values, accuracies_euclidean_weights, marker='o', linestyle='dashed', label='Euclidiana com Pesos')
plt.plot(k_values, accuracies_manhattan, marker='o', label='Manhattan')
plt.plot(k_values, accuracies_manhattan_weights, marker='o', linestyle=':', label='Manhattan com Pesos')

plt.xlabel('Número de Vizinhos k')
plt.ylabel('Acurácia')
plt.title('Acurácia do k-NN Variando o Valor de k e Medidas de Distância')
plt.legend()
plt.grid(True)
plt.show()
```
